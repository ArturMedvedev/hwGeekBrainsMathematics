```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 20, 201)

k_1 = 1
k_2 = 3

fig, ax = plt.subplots(nrows=1, ncols=1)

fig.set_size_inches(15, 6)
fig.subplots_adjust(wspace=1, hspace=1)

plt.plot(x, np.cos(k_1*x), color='red')
plt.plot(x, np.cos(k_2*x), color='black')

plt.title(' y(cos(k*x))')
plt.xlabel(' x')
plt.ylabel(' y')
plt.axis([0, 20, -1.5, 1.5])
plt.show()
```
